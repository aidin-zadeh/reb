```python
import os, inspect, sys
import json
from pprint import pprint

CURRENT_DIR = os.path.dirname(inspect.getabsfile(inspect.currentframe()))
ROOT_DIR = os.path.dirname(CURRENT_DIR)
sys.path.insert(0, ROOT_DIR)

from reb.src import pynyt
from reb.conf import APIKEY_NYT_ARTICLE
```

```python
section_name_values = ["business", "Business Day", "Real Estate"]
new_desk_values = ["Business", "Business Day", "Financial"]
type_of_material_values = ["News"]

for year in range(1960, 1980):
    for month in range(1, 13):
        print(f"--> Loading NYT archives for {year}-{month}")
        fname = f"nyt-arch-{year}-{month}.json"
        path = os.path.join(ROOT_DIR, "reb", "data", "raw")
        ffname = os.path.join(path, fname)
        with open(ffname, "r") as fp:
            mnews = json.load(fp)
            news_monthly = {}
            for news in mnews["response"]["docs"]:
                    if ("section_names" in news.keys() and news["section_name"] in section_name_values) or \
                        ("news_desk" in news.keys() and news["news_desk"] in new_desk_values) and \
                        (news["type_of_material"] in type_of_material_values):
                        date = news["pub_date"][0:10]
                        if date in news_monthly:
                            news_monthly[date].append(news)
                        else:
                            news_monthly.update({date: [news]})
        print(f"--> Writing NYT Business news archives for {year}-{month}")       
        fname = f"nyt-archive-business-{year}-{month}.json"
        path = os.path.join(ROOT_DIR, "reb", "data", "raw")
        ffname = os.path.join(path, fname)
        with open(ffname, "w") as fp:
            json.dump(news_monthly, fp)
#             print(f"--> Writing to '{fname}' at {path}")
```
